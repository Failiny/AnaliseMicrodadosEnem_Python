*Analisando com parâmetros socioeconômicos as notas:*

```python
selecaosocioeconomico = ['NU_INSCRICAO', 'SG_UF_PROVA',
                         'NU_NOTA_MT', 'NU_NOTA_REDACAO',
                         'Q001', 'Q002']
```

```python
analisesocioeconomico = microdadosEnem.filter(items=selecaosocioeconomico)
```

```python
analisesocioeconomico.head()
```

```python
analisesocioeconomico = analisesocioeconomico.dropna()
```

```python
analisesocioeconomico.head()
```

*Criando um dicionário - Descrição de cada alternativa assinalada na prova.*

**Q001: Até que série seu pai, ou homem responsável por você, estudou?
|| Q002: Até que série sua mãe, ou mulher responsável por você, estudou?**

```python
Q001e002Dicionario = {
    'A' : 'Nunca estudou',
    'B' : 'Não completou a 4 série / 5 ano do Ensino Fundamental',
    'C' : 'Completou a 4 série / 5 ano, mas não completou a 8 série / 9 ano do Ensino Fundamental',
    'D' : 'Completou a 8 serie / 9 ano do Ensino Fundamental, mas não completou Ensino Médio',
    'E' : 'COmpletou Ensino Médio, mas não completou a Faculdade',
    'F' : 'Completou a Faculdade, mas não completou a pós-graduação',
    'G' : 'Completou a pós-graduação',
    'H' : 'Não sei'}
```

```python
analisesocioeconomico.filter(items=['Q001','NU_INSCRICAO' ]).groupby('Q001').count()
```

*Criando uma coluna para atribuir o dicionário:*

```python
analisesocioeconomico['NO_Q001'] = [Q001e002Dicionario[resp] for resp in analisesocioeconomico.Q001]
```

```python
analisesocioeconomico.head()
```

```python
analisesocioeconomico['NO_Q002'] = [Q001e002Dicionario[resp] for resp in analisesocioeconomico.Q002]
```

```python
analisesocioeconomico.head()
```

*A sequencia abaixo estamos realizando a análise com a referência da escolaridade do Pai e da Mãe:*

```python
analisesocioeconomico.filter(items=['NU_INSCRICAO' , 'NO_Q001']).groupby('NO_Q001').count().sort_values(by='NU_INSCRICAO' , ascending=False)
```

```python
analisesocioeconomico.filter(items=['NU_INSCRICAO' , 'NO_Q002']).groupby('NO_Q002').count().sort_values(by='NU_INSCRICAO' , ascending=False)
```

```python
analisesocioeconomico.filter(items=['NU_NOTA_MT' , 'NO_Q001']).groupby('NO_Q001').mean().sort_values(by='NU_NOTA_MT' , ascending=False)
```

```python
analisesocioeconomico.filter(items=['NU_NOTA_MT' , 'NO_Q002']).groupby('NO_Q002').mean().sort_values(by='NU_NOTA_MT' , ascending=False)
```

```python
analisesocioeconomico.filter(items=['NU_NOTA_REDACAO' , 'NO_Q001']).groupby('NO_Q001').mean().sort_values(by='NU_NOTA_REDACAO' , ascending=False)
```

```python
analisesocioeconomico.filter(items=['NU_NOTA_REDACAO' , 'NO_Q002']).groupby('NO_Q002').mean().sort_values(by='NU_NOTA_REDACAO' , ascending=False)
```

```python
analisesocioeconomico.filter(items=['NU_NOTA_REDACAO','NO_Q001']).groupby('NO_Q001').describe()
```

*Analise abaixo com referência por estado:*

```python
analisesocioeconomico.filter(items=['NU_NOTA_REDACAO' , 'NO_Q001'])\
.where(analisesocioeconomico.SG_UF_PROVA == 'SP')\
.groupby('NO_Q001').mean().sort_values(by='NU_NOTA_REDACAO' , ascending=False)
```

```python
analisesocioeconomico.filter(items=['NU_NOTA_REDACAO' , 'NO_Q001'])\
.where(analisesocioeconomico.SG_UF_PROVA == 'RJ')\
.groupby('NO_Q001').mean().sort_values(by='NU_NOTA_REDACAO' , ascending=False)
```

*Fizemos um gráfico explorando a nota da redação pela escolaridade e por estado efetuado a prova. Aprendendo a manusear o matplotlib e algumas de suas funcionalidades:*

```python
#fig, ax = plt.subplots(figsize=(20,10))
#plt.suptitle('Nota Redação x Escolaridade x Estado')

ax = analisesocioeconomico.filter(items=['SG_UF_PROVA' , 'Q002' , 'NU_NOTA_REDACAO'])\
.groupby(['Q002' , 'SG_UF_PROVA']).mean().sort_values(by='NU_NOTA_REDACAO' , ascending=False)\
.unstack().plot(figsize=(16,8))
ax.set_title('Nota Redação x Escolaridade x Estado', fontsize=15)
#ax.legend(bbox_to_anchor=(1.01, 1), loc=0, borderaxespad=0.)

handles, labels = ax.get_legend_handles_labels()

import re
edited_labels = [re.search( ',\s(.+?)\)' , label).group(1) for label in labels]
ax.legend(edited_labels, bbox_to_anchor=(1.01, 1), loc=0, borderaxespad=0.)

textdictQ002 = ''
for key, value in Q001e002Dicionario.items():
  textdictQ002 = textdictQ002 + "{k} : {v}\n".format(k=key, v=value)

ax.text(0.05,0.7, textdictQ002, transform=ax.transAxes, fontsize=10)
```
